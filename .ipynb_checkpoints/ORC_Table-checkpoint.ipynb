#Filter


```python
def slope_filter(lines, thresh = 0.1, max_slope = 40):
  # Find slope = (y2-y1)/(x2-x1)
  slopes = np.apply_along_axis(lambda x: (x[3]-x[1])/(x[2]-x[0]),2, lines)
  verLines = []
  horLines = []
  for i, slope in enumerate(slopes):
    # Get vertical lines
    if (abs(slope) > max_slope) or (math.isinf(slope)):
      verLines.extend([lines[i]]) 
    # Get horizontal lines
    elif (abs(slope) < thresh):
      horLines.extend([lines[i]]) 

  return np.array(verLines), np.array(horLines)

def edge_filter(lines, width, height, thresh=50):
  lines_out = []
  for i, line in enumerate(lines):
    line = line[0]
    avg_x = (line[0] + line[2])/2
    avg_y = (line[1] + line[3])/2

    if (thresh < avg_x < (width-thresh)) and (thresh < avg_y < (height-thresh)):
      lines_out.extend([line])
      
  return np.array(lines_out)
```

```python
# Merge lines that near and has the same slope
def merge_lines(verLines, horLines, thresh_distance, thresh_gap=10):
  # Merge Verticle lines
  ## Cluster the near lines by [y0, avg(x)] 
  ## (y0 can be different base on the length of the detected line)
  verVector = np.apply_along_axis(lambda x: [x[1], x[3], (x[2]+x[0])/2], 2, verLines)
  horVector = np.apply_along_axis(lambda x: [x[0], x[2], (x[3]+x[1])/2], 2, horLines)

  print("verticle")
  verLines = merger(verVector, thresh_distance, thresh_gap, True)
  print("horizontal")
  horLines = merger(horVector, thresh_distance*20, 30)

  return verLines, horLines

def merger(vector, thresh_dis, thresh_gap, isVer=False):
  merged_vec = []
  merged_marker = []

  for i, v in enumerate(vector):
    v = v[0]
    merged = False
    for j in range(i+1, len(vector)):
      if j in merged_marker:
        continue

      v_temp = vector[j][0]
      diff = [abs(v[0]-v_temp[0]), abs(v[2]-v_temp[2])]
      print(diff)
      if (diff[1] < thresh_gap) \
      and (diff[0] < thresh_dis):
        v = [min(v[0], v_temp[0]), 
              max(v[1], v_temp[1]),
             min(v[2],v_temp[2])]
        merged_marker.append(j)
        merged = True
        
    if merged:
      if isVer:
        v = np.array([v[-1], v[0], v[-1], v[1]])
      else:
        v = np.array([v[0], v[-1], v[1], v[-1]])

      merged_vec.extend([[v]])
      
  return np.array(merged_vec)
```

```python
# Draw segment
def draw(img, lines):
  lines = lines.astype(int)
  for line in lines:
    line = line[0]
    print(line)
    cv2.line(~img, (line[0], line[1]), (line[2], line[3]), (255,0,0), 3)
  return ~img
```

#Detect Line

```python
import math

def line_detect(img, isTemplate=False, debug=False):
  # read_img = cv2.imread("/content/3B8283B2-3.jpg")
  # read_img = cv2.imread("/content/CD573519-2.png")
  img_temp = cv2.cvtColor(img.copy(), cv2.COLOR_BGR2GRAY)

  # img = cv2.dilate(img, None, iterations=2)
  # cv2_imshow(img)
  #Create default parametrization LSD
  lsd = cv2.ximgproc.createFastLineDetector()
  #Detect lines in the image
  lines = lsd.detect(img_temp)
  # print(lines)
  mylines = []
  for i in range(len(lines)):
    x1, y1, x2, y2 = lines[i][0]
    length = math.sqrt((x1-x2)**2 + (y1-y2)**2)
    # Filter length
    if length < img_temp.shape[0]/30:
      continue
    mylines.append(lines[i])

  # Draw detected lines in the image
  mylines = np.array(mylines)
  # Filter
  verLines, horLines = slope_filter(mylines)

  verLines, horLines = merge_lines(verLines, horLines, thresh_distance=img_temp.shape[0]/5)
  # lines = edge_filter(slopes, img.shape[1], img.shape[0], 50)

  # Draw  
  lines = np.concatenate((verLines, horLines), axis=0)
  # print(lines)
  # Draw on original img
  # drawn_img = lsd.drawSegments(img, lines)
  # drawn_img = draw(img, lines)
  if debug:
    # Take line only
    empty = np.zeros_like(img_temp)
    drawn_img = draw(empty, lines)

    return drawn_img
  else:
    drawn_img = draw(img, lines)

  if isTemplate:
    x_max, x_min = int(np.max(lines[:, [0,2]])), int(np.min(lines[:, [0,2]]))
    y_max, y_min = int(np.max(lines[:, [1,3]])), int(np.min(lines[:, [1,3]]))

    drawn_img = drawn_img[y_min:y_max, x_min:x_max]
    return drawn_img
  
  # drawn_img = cv2.dilate(drawn_img, None, iterations=2)
  # cv2_imshow(drawn_img)
  return drawn_img
```
